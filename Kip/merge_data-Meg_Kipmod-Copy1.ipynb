```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file1 = "data/life-expectancy-at-birth-total-years.csv"
file2 = "data/median-age.csv"
file3 = "data/population-density-vs-prosperity.csv"
file4 = "data/urban-and-rural-population.csv"
file6 = "data/urban-vs-rural-majority.csv"

# Read File and store into Pandas data frame
life_expectancy = pd.read_csv(file1)
median_age = pd.read_csv(file2)
density_vs_prosperity = pd.read_csv(file3)
urban_rural_population = pd.read_csv(file4)
urban_vs_rural_majority = pd.read_csv(file6)
```

```python
density_vs_prosperity.shape
```

```python
urban_rural_population.shape
```

```python
urban_vs_rural_majority.shape
```

```python
# Merge data
merged_data = pd.merge(density_vs_prosperity, urban_rural_population, on=["Entity", "Code", "Year"])
merged_data.shape
```

```python
merged_data = pd.merge(merged_data, urban_vs_rural_majority, on=["Entity", "Code", "Year"])
merged_data.shape
```

```python
# Export CSV
merged_data.to_csv('merged_data_noslums_compact.csv') 

merged_data.head()
```

```python
merged_data.tail()
```

```python
#Format columns
merged_data = merged_data.rename(columns={"Population density (people per km² of land area)": "pop_density",
                                          "GDP per capita (constant 2011 international $)" : "gdp",
                                         "Total population (Gapminder)": "total_pop",
                                         "Urban population": "urban_pop",
                                         "Rural population": "rural_pop",
                                         "Urban (%)":"urban_pct",
                                         "Rural (%)": "rural_pct"})
merged_data.head()
```

```python
# Filter by most recent year
merged_data = merged_data[merged_data["Year"] == 2014]
merged_data.head()
```

```python
# Export CSV
merged_data.to_csv('merged_data.csv')  
```
